```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Define Dataset
x = np.array([2, 4, 6, 8, 10])
y = np.array([5, 11, 13, 19, 22])
```

```python
# 2. Calculate Means
mean_x = np.mean(x)
mean_y = np.mean(y)
```

```python
# 3. Calculate Deviations and Products
dx = x - mean_x
dy = y - mean_y
dx_dy = dx * dy
dx_sq = dx**2
```

```python
# 4. Create Calculation Table
df = pd.DataFrame({
    'x': x,
    'y': y,
    'dx (x - mean_x)': dx,
    'dy (y - mean_y)': dy,
    'dx * dy': dx_dy,
    'dx^2': dx_sq
})

# Add Sum Row
df.loc['Sum'] = df.sum()
```

```python
# 5. Print Results
print("--- Linear Regression Calculation Table ---")
print(df.to_string(index=True))
print(f"\nMean of x: {mean_x}")
print(f"Mean of y: {mean_y}")
print(f"Mean Deviation of x: {np.mean(np.abs(dx))}")
print(f"Mean Deviation of y: {np.mean(np.abs(dy))}")
```

```python
# 6. Calculate Regression Equation (y = mx + c)
# m = Σ(dx*dy) / Σ(dx^2)
m = df.loc['Sum', 'dx * dy'] / df.loc['Sum', 'dx^2']
c = mean_y - (m * mean_x)
```

```python
# 7. Visualization
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, m*x + c, color='red', label=f'Regression Line: y = {m:.2f}x + {c:.2f}')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Linear Regression: Deviations and Best Fit Line')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Define Dataset
x = np.array([2, 4, 6, 8, 10])
y = np.array([5, 11, 13, 19, 22])

# 2. Calculate Means
mean_x = np.mean(x)
mean_y = np.mean(y)

# 3. Calculate Deviations and Products
dx = x - mean_x
dy = y - mean_y
dx_dy = dx * dy
dx_sq = dx**2

# 4. Create Calculation Table
df = pd.DataFrame({
    'x': x,
    'y': y,
    'dx (x - mean_x)': dx,
    'dy (y - mean_y)': dy,
    'dx * dy': dx_dy,
    'dx^2': dx_sq
})

# Add Sum Row
df.loc['Sum'] = df.sum()

# 5. Print Results
print("--- Linear Regression Calculation Table ---")
print(df.to_string(index=True))
print(f"\nMean of x: {mean_x}")
print(f"Mean of y: {mean_y}")
print(f"Mean Deviation of x: {np.mean(np.abs(dx))}")
print(f"Mean Deviation of y: {np.mean(np.abs(dy))}")

# 6. Calculate Regression Equation (y = mx + c)
# m = Σ(dx*dy) / Σ(dx^2)
m = df.loc['Sum', 'dx * dy'] / df.loc['Sum', 'dx^2']
c = mean_y - (m * mean_x)

# 7. Visualization
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, m*x + c, color='red', label=f'Regression Line: y = {m:.2f}x + {c:.2f}')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Linear Regression: Deviations and Best Fit Line')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
